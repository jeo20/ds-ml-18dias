```python
df_combinado =  pd.concat([df5, df6, df7, df8, df9, df10], ignore_index=True)
df_combinado.info()
```

AGREGO COLUMNAS IMPORTE

```python
# Crear una copia del DataFrame df_combinado2 para evitar modificar el original
df_nuevo = df_combinado.copy()

# AGREGO OCTUBRE
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE OCTUBRE", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 10, "IMPORTE OCTUBRE"] = df_nuevo.loc[df_nuevo["PERIODO"] == 10, "IMPORTE 283"]

# AGREGO SEPTIEMBRE
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE SEPTIEMBRE", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 9, "IMPORTE SEPTIEMBRE"] = df_nuevo.loc[df_nuevo["PERIODO"] == 9, "IMPORTE 283"]


# AGREGO AGOSTO
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE AGOSTO", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 8, "IMPORTE AGOSTO"] = df_nuevo.loc[df_nuevo["PERIODO"] == 8, "IMPORTE 283"]

# AGREGO JULIO
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE JULIO", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 7, "IMPORTE JULIO"] = df_nuevo.loc[df_nuevo["PERIODO"] == 7, "IMPORTE 283"]


# AGREGO JUNIO
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE JUNIO", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 6, "IMPORTE JUNIO"] = df_nuevo.loc[df_nuevo["PERIODO"] == 6, "IMPORTE 283"]


# AGREGO MAYO
df_nuevo.insert(df_nuevo.columns.get_loc("IMPORTE 283") + 1, "IMPORTE MAYO", 0)
df_nuevo.loc[df_nuevo["PERIODO"] == 5, "IMPORTE MAYO"] = df_nuevo.loc[df_nuevo["PERIODO"] == 5, "IMPORTE 283"]

# Mostrar el nuevo DataFrame
df_nuevo
```

```python
df_nuevo['IMPORTE MAYO'] = df_nuevo['IMPORTE MAYO'].astype(float)
df_nuevo['IMPORTE JUNIO'] = df_nuevo['IMPORTE JUNIO'].astype(float)
df_nuevo['IMPORTE JULIO'] = df_nuevo['IMPORTE JULIO'].astype(float)
df_nuevo['IMPORTE AGOSTO'] = df_nuevo['IMPORTE AGOSTO'].astype(float)
df_nuevo['IMPORTE SEPTIEMBRE'] = df_nuevo['IMPORTE SEPTIEMBRE'].astype(float)
df_nuevo['IMPORTE OCTUBRE'] = df_nuevo['IMPORTE OCTUBRE'].astype(float)
df_nuevo.info()
```

```python
# Agrupar por legajo y obtener el mÃ¡ximo importe para cada periodo
df_agrupado = df_nuevo.groupby('LEGAJO').agg({
    'IMPORTE MAYO': 'max',
    'IMPORTE JUNIO': 'max',
    'IMPORTE JULIO': 'max',
    'IMPORTE AGOSTO': 'max',
    'IMPORTE SEPTIEMBRE': 'max',
    'IMPORTE OCTUBRE': 'max',
    'APELLIDO': 'first', 
    'CARGO': 'first',  
    'PLANTA': 'first', 
    'CATEGORIA': 'first',
    'ANEXO': 'first',
    'ITEM': 'first',
    'AGRUPAMIENTO': 'first'
})


# Resetear el Ã­ndice para que el legajo sea una columna nuevamente
df_agrupado = df_agrupado.reset_index()

# Eliminar las columnas que ya no se necesitan en el DataFrame original
df_sin_duplicados = df_nuevo.drop_duplicates(subset=['LEGAJO'])

# Fusionar el DataFrame con las columnas del importe mÃ¡ximo y el DataFrame sin duplicados
df_final = pd.merge(df_sin_duplicados, df_agrupado[['LEGAJO', 'IMPORTE MAYO', 'IMPORTE JUNIO',
                                                   'IMPORTE JULIO', 'IMPORTE AGOSTO',
                                                   'IMPORTE SEPTIEMBRE', 'IMPORTE OCTUBRE']],
                     on='LEGAJO', how='left')

# Mostrar el DataFrame final
df_final
```
