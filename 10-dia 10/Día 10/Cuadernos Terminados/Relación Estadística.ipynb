# Relación Estadística

No existe una forma universalmente perfecta de visualizar los datos. Cuando analizamos datos nos hacemos diferentes preguntas, y cada pregunta se responde con diferentes gráficos. Para eso, Seaborn facilita el cambio entre diferentes gráficos, ya que su sintaxis está orientada a conjuntos de datos.

Te voy a presentar estas 3 funciones generales de Seaborn que sirven para crear gráficos. Pero lo interesante es la flexibilidad de estas funciones ya que cada una de ellas permite crear diferentes tipos de gráficos para el mismo conjunto de datos.

![CUADRO](funciones_seaborn.png)

Entonces `relplot()`, que es la función que usamos en la lección anterior para mostrarte cómo funciona Seaborn, nos permite crear tanto **gráficos scatter** (o de dispersión) o **gráficos de línea**, y luego tenemos a `displot()` y `catplot()`, que cada una de ellas puede generar los gráficos que ves detallados en la parte inferior.

Estas tres funciones no están divididas caprichosamente, sino que cada una de ellas representa tres grandes categorías o modelos de la estadística:
* Relplot() para el modelo de relación estadística
* Displot() para las representaciones distributivas
* catplot() para los gráficos de datos categóricos

No te preocupes si no sabes qué significan estos nombres extraños, porque vamos a ver a cada uno de ellos por separado, y en esta lección comenzamos con la **Relación Estadística**, y su función `relplot()`.

La **relación estadística** se refiere a la búsqueda de comprender *cómo las variables se relacionan entre sí*, se refiere al vínculo o a la conexión que existe entre dos o más variables dentro de un conjunto de datos. Este vínculo puede manifestarse de diversas formas, como tendencias, correlaciones o patrones predictivos. En esta lección, nos vamos a sumergir en el concepto de **relación estadística** y vamos a explorar cómo Seaborn nos ayuda a visualizar y entender estas relaciones.


#### ¿Qué es la Relación Estadística?
Cuando existe una relación estadística entre variables, esto indica que **los cambios en una variable** están asociados con **cambios en otra variable**. Estas relaciones pueden ser *positivas* (ambas variables se mueven en la misma dirección), *negativas* (cuando las variables se mueven en direcciones opuestas) o incluso *no lineales*, donde la relación no sigue una línea recta.


#### Visualizando Relaciones Estadísticas con Seaborn
Seaborn simplifica el proceso de exploración y visualización de las relaciones estadísticas con el uso de una función muy flexible, que ya hemos conocido en la lección anterior a modo de ejemplo de la sintaxis de Seaborn: la función `relplot()`

Vamos a cargar uno de los modelos de datos de ejemplo que nos ofrece Seaborn:

```python
import seaborn as sns
```

```python
propinas = sns.load_dataset("tips")
propinas
```

Aquí vemos registros de propinas basados en los datos:
* `total_bill` (factura total)
* `tip` (propina)
* `sex` (género)
* `smoker` (si es fumador o no)
* `day` (día de la semana)
* `time` (si fue almuerzo o cena)
* `size` (cantidad de personas).

Entonces yo ahora quiero que relplot me muestre la relación entre la **factura total** y la **propina** que han dejado.

```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip");
```

Vemos que Seaborn ha decidido que la mejor manera demostrar esta información es con un **gráfico de dispersión** (un gráfico scatter). Pero `relplot()`, como vimos en la imagen superior, puede generar tanto **gráficos scatter** como **gráficos de línea**.

Entonces podemos modificar esto gracias al atributo `kind`, que quiere decir "*tipo*" o "*clase*".

```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip",
           kind="line");
```

Y ahora los mismos datos **se representan con una línea**, aunque en este caso tal vez no es la mejor idea.

También podemos ver que `relplot()` nos permite agregar otros atributos para comparar diferencias.

```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip",
           hue="smoker",
           style="smoker",
           size="size",
           kind="line");
```

Bueno, esto se ve horrible, así que quitaré `kind` para que vuelva a ser un scatter… 

```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip",
           hue="smoker",
           style="smoker",
           size="size");
```


Y por ultimo quiero mostrarte un atributo muy interesante que es `col`, que significa *columna*, y que permite crear diferentes columnas para agrupar los datos según alguna otra categoría y comparar las diferencias.


```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip",
           hue="smoker",
           style="smoker",
           size="size",
           col="time");
```

Observa que todo el tiempo `relplot()` ha creado automáticamente una **tabla de referencias** a la derecha, aunque no se lo hemos pedido, y esa es otra de las ventajas de Seaborn, pero si quieres quitarla, simplemente cambias el valor del atributo `legend`.

```python
sns.relplot(data=propinas,
           x="total_bill",
           y="tip",
           hue="smoker",
           style="smoker",
           size="size",
           col="time",
           legend=None);
```

Entonces, comprender las **relaciones estadísticas entre variables** es un paso crucial en el análisis de datos. Y con herramientas visuales poderosas como `relplot()` de Seaborn, podemos descubrir fácilmente cómo las variables interactúan entre sí, lo cual es esencial para realizar análisis más profundos.
