```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("D:/Cursos/ds_ml_18dias/09-dia 9/Día 9/32.1 Iluminacion.csv")
df
```

```python
help(plt.hexbin)
```

```python
# plot:
fig, ax = plt.subplots()

hb = ax.hexbin(df['Latitud'], df['Longitud'], C=df['Luminosidad'], gridsize=50, cmap='Greens')
ax.set_facecolor('black')
plt.colorbar(hb, label='nivel de sombra')
plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Luminosidad en Villa Luz');
```
